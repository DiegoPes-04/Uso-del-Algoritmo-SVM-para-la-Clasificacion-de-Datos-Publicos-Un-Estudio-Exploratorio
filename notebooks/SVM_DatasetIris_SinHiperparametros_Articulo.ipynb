```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Crear un DataFrame con las cargas de las variables
cargas = pd.DataFrame(
    pca.components_.T,
    index=X.columns,
    columns=[f'PC{i+1}' for i in range(pca.n_components_)]
)

# Calcular la importancia total de cada variable (usando la suma del valor absoluto de sus cargas)
importancia = cargas.abs().sum(axis=1).sort_values(ascending=False)

# Graficar
plt.figure(figsize=(8,5))
plt.bar(importancia.index, importancia.values)
plt.title('Importancia de las variables según PCA')
plt.xlabel('Variables')
plt.ylabel('Importancia (suma de cargas absolutas)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Si quieres ver también los valores exactos:
print(importancia)
```

ENTRENAMIENTO DEL MODELO

```python
from sklearn.model_selection import cross_val_score, StratifiedKFold
from sklearn import svm
from sklearn.metrics import confusion_matrix, classification_report
from mlxtend.plotting import plot_confusion_matrix
import numpy as np
import time
import matplotlib.pyplot as plt

# Features (X) y etiquetas (y)
X = datosIris.drop(['Species'], axis=1)
y = datosIris['Species']

# Lista de kernels a evaluar
kernels = ['linear', 'rbf', 'poly', 'sigmoid']

# Diccionarios para guardar resultados
resultados_cv = {}
tiempos_promedio = {}

# Definir validación cruzada estratificada
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

for kernel in kernels:
    clf = svm.SVC(kernel=kernel)
    tiempos = []
    scores = []

    for train_index, test_index in cv.split(X, y):
        X_train, X_test = X.iloc[train_index], X.iloc[test_index]
        y_train, y_test = y.iloc[train_index], y.iloc[test_index]

        t0 = time.time()
        clf.fit(X_train, y_train)
        t1 = time.time()

        tiempos.append(t1 - t0)
        scores.append(clf.score(X_test, y_test))

    resultados_cv[kernel] = {
        'media': np.mean(scores),
        'desviacion': np.std(scores),
        'valores_individuales': scores
    }
    tiempos_promedio[kernel] = np.mean(tiempos)

    print(f"\n=== Kernel: {kernel} ===")
    print(f"Scores individuales: {scores}")
    print(f"Precisión promedio: {np.mean(scores):.4f}")
    print(f"Desviación estándar: {np.std(scores):.4f}")
    print(f"Tiempo promedio de entrenamiento: {np.mean(tiempos):.6f} s")

    #Entrenar una sola vez con todo el dataset para generar matriz y reporte
    clf.fit(X, y)
    y_pred = clf.predict(X)

    matriz = confusion_matrix(y, y_pred)
    print("\nMatriz de confusión:")
    print(matriz)

    plot_confusion_matrix(conf_mat=matriz, figsize=(5, 5))
    plt.title(f"Matriz de confusión - Kernel: {kernel}")
    plt.show()

    print("Classification Report:")
    print(classification_report(y, y_pred))

# ==============================
# Visualización de precisión
# ==============================
nombres = list(resultados_cv.keys())
medias = [resultados_cv[k]['media'] for k in nombres]
desviaciones = [resultados_cv[k]['desviacion'] for k in nombres]

plt.figure(figsize=(8,5))
plt.bar(nombres, medias, yerr=desviaciones, capsize=5,
        color=['skyblue', 'lightgreen', 'salmon', 'orange'])
plt.ylim(0, 1.1)
plt.xlabel('Kernel')
plt.ylabel('Precisión promedio (CV=5)')
plt.title('Comparación de precisión promedio por kernel')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()

# ==============================
# Visualización de tiempos
# ==============================
tiempos = [tiempos_promedio[k] for k in nombres]

plt.figure(figsize=(8,5))
plt.bar(nombres, tiempos, color=['skyblue', 'lightgreen', 'salmon', 'orange'])
plt.xlabel('Kernel')
plt.ylabel('Tiempo promedio de entrenamiento (s)')
plt.title('Comparación de tiempos promedio de entrenamiento por kernel')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()


#1. Encontrar el kernel con mejor precisión
mejor_kernel = max(resultados_cv, key=lambda k: resultados_cv[k]['media'])
print(f"\nEl mejor kernel es: {mejor_kernel} con precisión de {resultados_cv[mejor_kernel]['media']:.4f}")

#2. Entrenar modelo final con ese kernel
mejor_modelo = svm.SVC(kernel=mejor_kernel)
mejor_modelo.fit(X, y)

#3. Definir el mapeo de clases numéricas a nombres de especie
mapa_clases = {
    1: 'Iris-setosa',
    2: 'Iris-versicolor',
    3: 'Iris-virginica'
}

#4. Probar el modelo con nuevos datos
nueva_muestra = [[5.9, 3, 5.1, 1.8]]
prediccion = mejor_modelo.predict(nueva_muestra)[0]
especie = mapa_clases[prediccion]

print(f"\nPredicción numérica: {prediccion}")
print(f"Especie predicha: {especie}")
```

```python
from sklearn.preprocessing import label_binarize
from sklearn.metrics import roc_curve, auc
from itertools import cycle

# 1. Binarizamos las clases (1,2,3 → vectores One-Hot)
y_bin = label_binarize(y, classes=[1, 2, 3])
n_clases = y_bin.shape[1]

# 2. Dividir datos en train/test para evaluación ROC
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y_bin, test_size=0.3, random_state=42, stratify=y)

# 3. Evaluar cada kernel
for kernel in ['linear', 'rbf', 'poly', 'sigmoid']:
    clf = svm.SVC(kernel=kernel, probability=True)  # ✅ activamos probabilidades
    clf.fit(X_train, y_train.argmax(axis=1))
    y_score = clf.predict_proba(X_test)

    # 4. Calcular ROC y AUC por clase
    fpr = dict()
    tpr = dict()
    roc_auc = dict()
    for i in range(n_clases):
        fpr[i], tpr[i], _ = roc_curve(y_test[:, i], y_score[:, i])
        roc_auc[i] = auc(fpr[i], tpr[i])

    # 5. Graficar ROC para cada clase
    plt.figure()
    colors = cycle(['blue', 'red', 'green'])
    for i, color in zip(range(n_clases), colors):
        plt.plot(fpr[i], tpr[i], color=color, lw=2,
                 label=f'Clase {i+1} (AUC = {roc_auc[i]:0.2f})')
    plt.plot([0, 1], [0, 1], 'k--', lw=2)
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('Tasa de Falsos Positivos (FPR)')
    plt.ylabel('Tasa de Verdaderos Positivos (TPR)')
    plt.title(f'Curva ROC - Kernel: {kernel}')
    plt.legend(loc="lower right")
    plt.show()
```
