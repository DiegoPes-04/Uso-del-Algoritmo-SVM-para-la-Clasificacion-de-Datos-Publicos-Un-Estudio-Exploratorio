```python
"""
SVM en el dataset Iris: estudio exploratorio y guía reproducible

"""

# ============================================================================
# CONFIGURACIÓN INICIAL Y LIBRERÍAS
# ============================================================================
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import (train_test_split, GridSearchCV,
                                     StratifiedKFold, cross_val_score,
                                     validation_curve, learning_curve)
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (classification_report, confusion_matrix,
                             accuracy_score, precision_score, recall_score,
                             f1_score, roc_auc_score, roc_curve, auc)
from sklearn.preprocessing import label_binarize
import warnings
warnings.filterwarnings('ignore')

# Configuración de estilo
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)

# Semilla para reproducibilidad
RANDOM_STATE = 42
np.random.seed(RANDOM_STATE)

print("="*80)
print("SVM EN DATASET IRIS - METODOLOGÍA OSEMN")
print("="*80)

# ============================================================================
# FASE 1: SELECCIÓN DE DATOS
# ============================================================================
print("\n[FASE 1: SELECCIÓN DE DATOS]")
print("-" * 80)

# Cargar dataset
datosIris = pd.read_csv("Iris.csv")
print(f"✓ Dataset cargado: {datosIris.shape[0]} muestras, {datosIris.shape[1]} columnas")
print(f"✓ Columnas: {list(datosIris.columns)}")

# ============================================================================
# FASE 2: PREPROCESAMIENTO Y LIMPIEZA
# ============================================================================
print("\n[FASE 2: PREPROCESAMIENTO Y LIMPIEZA]")
print("-" * 80)

# Eliminar columna Id (no aporta información)
datosIris = datosIris.drop(columns=["Id"], errors='ignore')
print("✓ Columna 'Id' eliminada")

# Verificar valores nulos
print(f"✓ Valores nulos: {datosIris.isnull().sum().sum()}")

# Verificar tipos de datos
print("\n✓ Información del dataset:")
print(datosIris.info())

# Balance de clases
print("\n✓ Balance de clases:")
print(datosIris['Species'].value_counts())

# Visualización del balance
plt.figure(figsize=(8, 5))
sns.countplot(data=datosIris, x='Species', palette='viridis')
plt.title('Distribución de Especies (Balance de Clases)', fontsize=14, fontweight='bold')
plt.xlabel('Especie')
plt.ylabel('Frecuencia')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# ============================================================================
# FASE 3: TRANSFORMACIÓN Y ANÁLISIS EXPLORATORIO (EDA)
# ============================================================================
print("\n[FASE 3: TRANSFORMACIÓN Y ANÁLISIS EXPLORATORIO]")
print("-" * 80)

# Estadísticas descriptivas
print("\n✓ Estadísticas descriptivas:")
print(datosIris.describe())

# Correlación entre features
features_numericas = datosIris.select_dtypes(include=[np.number])
corr_matrix = features_numericas.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm",
            square=True, linewidths=0.5)
plt.title('Matriz de Correlación entre Features', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()

# Pairplot para visualizar relaciones
print("\n✓ Generando pairplot...")
sns.pairplot(datosIris, hue="Species", diag_kind="hist", palette="viridis")
plt.suptitle('Pairplot del Dataset Iris', y=1.02, fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()

# Distribuciones univariadas (histogramas)
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
features = features_numericas.columns
for i, feature in enumerate(features):
    ax = axes[i//2, i%2]
    ax.hist(datosIris[feature], bins=20, color='skyblue', edgecolor='black', alpha=0.7)
    ax.set_title(f'Distribución de {feature}', fontweight='bold')
    ax.set_xlabel(feature)
    ax.set_ylabel('Frecuencia')
plt.tight_layout()
plt.show()

# Boxplots para detectar outliers
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
for i, feature in enumerate(features):
    ax = axes[i//2, i%2]
    sns.boxplot(y=datosIris[feature], ax=ax, color='lightgreen')
    ax.set_title(f'Boxplot de {feature}', fontweight='bold')
plt.tight_layout()
plt.show()

# ============================================================================
# PREPARACIÓN DE DATOS PARA MODELADO
# ============================================================================
print("\n[PREPARACIÓN DE DATOS PARA MODELADO]")
print("-" * 80)

# Separar features (X) y target (y)
X = datosIris.drop(['Species'], axis=1)
y = datosIris['Species']

# Mapear especies a valores numéricos
especies_map = {"Iris-setosa": 0, "Iris-versicolor": 1, "Iris-virginica": 2}
y = y.map(especies_map)

print(f"✓ Features (X): {X.shape}")
print(f"✓ Target (y): {y.shape}")
print(f"✓ Clases: {np.unique(y)}")

# Split estratificado (70% train, 30% test)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=RANDOM_STATE, stratify=y
)

print(f"\n✓ Conjunto de entrenamiento: {X_train.shape[0]} muestras")
print(f"✓ Conjunto de prueba: {X_test.shape[0]} muestras")
print(f"✓ Distribución en train: {np.bincount(y_train)}")
print(f"✓ Distribución en test: {np.bincount(y_test)}")

# ============================================================================
# FASE 4: MINERÍA DE DATOS (DATA MINING) - MODELADO
# ============================================================================
print("\n[FASE 4: MINERÍA DE DATOS - MODELADO]")
print("="*80)

# ----------------------------------------------------------------------------
# 4.1: MODELO BASELINE - REGRESIÓN LOGÍSTICA
# ----------------------------------------------------------------------------
print("\n[4.1: MODELO BASELINE - REGRESIÓN LOGÍSTICA]")
print("-" * 80)

# Pipeline para baseline
pipeline_baseline = Pipeline([
    ('scaler', StandardScaler()),
    ('classifier', LogisticRegression(random_state=RANDOM_STATE, max_iter=1000))
])

# Validación cruzada estratificada
cv_strategy = StratifiedKFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE)
scores_baseline = cross_val_score(pipeline_baseline, X_train, y_train,
                                  cv=cv_strategy, scoring='accuracy')

# Entrenar modelo baseline
pipeline_baseline.fit(X_train, y_train)
y_pred_baseline = pipeline_baseline.predict(X_test)

# Métricas del baseline
acc_baseline = accuracy_score(y_test, y_pred_baseline)
precision_baseline = precision_score(y_test, y_pred_baseline, average='weighted')
recall_baseline = recall_score(y_test, y_pred_baseline, average='weighted')
f1_baseline = f1_score(y_test, y_pred_baseline, average='weighted')

print(f"\n✓ BASELINE - Regresión Logística:")
print(f"  - Accuracy (CV): {scores_baseline.mean():.4f} ± {scores_baseline.std():.4f}")
print(f"  - Accuracy (Test): {acc_baseline:.4f}")
print(f"  - Precision (Test): {precision_baseline:.4f}")
print(f"  - Recall (Test): {recall_baseline:.4f}")
print(f"  - F1-Score (Test): {f1_baseline:.4f}")

# ----------------------------------------------------------------------------
# 4.2: SVM CON DIFERENTES KERNELS Y GRIDSEARCH
# ----------------------------------------------------------------------------
print("\n[4.2: SVM CON DIFERENTES KERNELS]")
print("-" * 80)

import time

# Diccionarios para almacenar resultados de cada kernel
resultados_kernels = {}
tiempos_kernels = {}
mejores_params_kernels = {}

# Kernels a evaluar
kernels = ['linear', 'rbf', 'poly', 'sigmoid']

print("\n✓ Evaluando cada kernel individualmente con GridSearchCV...")
print(f"  - Validación cruzada: {cv_strategy.n_splits}-fold estratificada")
print(f"  - Métrica de optimización: accuracy")
print(f"  - Semilla aleatoria: {RANDOM_STATE}")

# Evaluar cada kernel por separado
for kernel in kernels:
    print(f"\n{'='*80}")
    print(f"EVALUANDO KERNEL: {kernel.upper()}")
    print('='*80)

    # Pipeline SVM
    pipeline_svm = Pipeline([
        ('scaler', StandardScaler()),
        ('classifier', SVC(kernel=kernel, random_state=RANDOM_STATE, probability=True))
    ])

    # Grid de hiperparámetros específico para cada kernel
    if kernel == 'linear':
        param_grid = {
            'classifier__C': [0.1, 1, 10, 100]
        }
    elif kernel == 'rbf':
        param_grid = {
            'classifier__C': [0.1, 1, 10, 100],
            'classifier__gamma': ['scale', 'auto', 0.001, 0.01, 0.1]
        }
    elif kernel == 'poly':
        param_grid = {
            'classifier__C': [0.1, 1, 10, 100],
            'classifier__gamma': ['scale', 'auto', 0.001, 0.01, 0.1],
            'classifier__degree': [2, 3, 4]
        }
    elif kernel == 'sigmoid':
        param_grid = {
            'classifier__C': [0.1, 1, 10, 100],
            'classifier__gamma': ['scale', 'auto', 0.001, 0.01, 0.1]
        }

    # GridSearchCV para este kernel
    grid_search = GridSearchCV(
        pipeline_svm,
        param_grid,
        cv=cv_strategy,
        scoring='accuracy',
        n_jobs=-1,
        verbose=0,
        return_train_score=True
    )

    # Entrenar y medir tiempo
    print(f"\n✓ Ejecutando GridSearchCV para kernel {kernel}...")
    tiempo_inicio = time.time()
    grid_search.fit(X_train, y_train)
    tiempo_fin = time.time()
    tiempo_total = tiempo_fin - tiempo_inicio

    # Guardar tiempo
    tiempos_kernels[kernel] = tiempo_total

    # Mejores hiperparámetros
    mejores_params_kernels[kernel] = grid_search.best_params_
    print(f"\n✓ Mejores hiperparámetros para {kernel}:")
    for param, value in grid_search.best_params_.items():
        print(f"  - {param}: {value}")

    # Modelo óptimo para este kernel
    best_model_kernel = grid_search.best_estimator_

    # Predicciones en test
    tiempo_inicio_test = time.time()
    y_pred_kernel = best_model_kernel.predict(X_test)
    y_proba_kernel = best_model_kernel.predict_proba(X_test)
    tiempo_fin_test = time.time()
    tiempo_test = tiempo_fin_test - tiempo_inicio_test

    # Calcular métricas
    acc = accuracy_score(y_test, y_pred_kernel)
    precision = precision_score(y_test, y_pred_kernel, average='weighted')
    recall = recall_score(y_test, y_pred_kernel, average='weighted')
    f1 = f1_score(y_test, y_pred_kernel, average='weighted')

    # AUC multiclase
    y_test_bin = label_binarize(y_test, classes=[0, 1, 2])
    auc_scores = []
    for i in range(3):
        auc_scores.append(roc_auc_score(y_test_bin[:, i], y_proba_kernel[:, i]))
    auc_macro = np.mean(auc_scores)

    # Guardar resultados
    resultados_kernels[kernel] = {
        'accuracy': acc,
        'precision': precision,
        'recall': recall,
        'f1_score': f1,
        'auc_macro': auc_macro,
        'cv_score': grid_search.best_score_,
        'cv_std': grid_search.cv_results_['std_test_score'][grid_search.best_index_],
        'tiempo_entrenamiento': tiempo_total,
        'tiempo_test': tiempo_test,
        'modelo': best_model_kernel,
        'y_pred': y_pred_kernel,
        'confusion_matrix': confusion_matrix(y_test, y_pred_kernel)
    }

    # Mostrar resultados
    print(f"\n✓ RESULTADOS PARA KERNEL {kernel.upper()}:")
    print(f"  - Tiempo de entrenamiento (GridSearch): {tiempo_total:.4f} segundos")
    print(f"  - Tiempo de predicción (test): {tiempo_test:.6f} segundos")
    print(f"  - Accuracy (CV): {grid_search.best_score_:.4f} ± {resultados_kernels[kernel]['cv_std']:.4f}")
    print(f"  - Accuracy (Test): {acc:.4f}")
    print(f"  - Precision (Test): {precision:.4f}")
    print(f"  - Recall (Test): {recall:.4f}")
    print(f"  - F1-Score (Test): {f1:.4f}")
    print(f"  - AUC Macro (Test): {auc_macro:.4f}")

    # Matriz de confusión
    cm = resultados_kernels[kernel]['confusion_matrix']
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=['Setosa', 'Versicolor', 'Virginica'],
                yticklabels=['Setosa', 'Versicolor', 'Virginica'])
    plt.title(f'Matriz de Confusión - Kernel: {kernel.upper()}', fontsize=14, fontweight='bold')
    plt.ylabel('Clase Real')
    plt.xlabel('Clase Predicha')
    plt.tight_layout()
    plt.show()

    # Classification report
    print(f"\n✓ REPORTE DE CLASIFICACIÓN - Kernel {kernel.upper()}:")
    target_names = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']
    print(classification_report(y_test, y_pred_kernel, target_names=target_names))

# ----------------------------------------------------------------------------
# COMPARACIÓN DE TODOS LOS KERNELS
# ----------------------------------------------------------------------------
print("\n" + "="*80)
print("COMPARACIÓN DE TODOS LOS KERNELS")
print("="*80)

# Tabla comparativa
print("\n" + "-"*120)
print(f"{'Kernel':<12} {'Acc (CV)':<12} {'Acc (Test)':<12} {'Precision':<12} {'Recall':<12} {'F1-Score':<12} {'AUC':<12} {'T.Train(s)':<12} {'T.Test(s)':<12}")
print("-"*120)
for kernel in kernels:
    r = resultados_kernels[kernel]
    print(f"{kernel:<12} {r['cv_score']:<12.4f} {r['accuracy']:<12.4f} {r['precision']:<12.4f} "
          f"{r['recall']:<12.4f} {r['f1_score']:<12.4f} {r['auc_macro']:<12.4f} "
          f"{r['tiempo_entrenamiento']:<12.4f} {r['tiempo_test']:<12.6f}")
print("-"*120)

# Visualización comparativa de métricas
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
fig.suptitle('Comparación de Métricas por Kernel', fontsize=16, fontweight='bold')

metricas = ['accuracy', 'precision', 'recall', 'f1_score', 'auc_macro']
titulos = ['Accuracy (Test)', 'Precision', 'Recall', 'F1-Score', 'AUC Macro']
colores = ['skyblue', 'lightgreen', 'salmon', 'orange']

for idx, (metrica, titulo) in enumerate(zip(metricas, titulos)):
    ax = axes[idx // 3, idx % 3]
    valores = [resultados_kernels[k][metrica] for k in kernels]
    barras = ax.bar(kernels, valores, color=colores)
    ax.set_title(titulo, fontweight='bold')
    ax.set_ylabel('Score')
    ax.set_ylim([0, 1.1])
    ax.grid(axis='y', alpha=0.3)
    # Añadir valores encima de las barras
    for barra in barras:
        height = barra.get_height()
        ax.text(barra.get_x() + barra.get_width()/2., height,
                f'{height:.3f}', ha='center', va='bottom', fontsize=9)

# Gráfico de tiempos
ax = axes[1, 2]
tiempos_train = [resultados_kernels[k]['tiempo_entrenamiento'] for k in kernels]
barras = ax.bar(kernels, tiempos_train, color=colores)
ax.set_title('Tiempo de Entrenamiento', fontweight='bold')
ax.set_ylabel('Segundos')
ax.grid(axis='y', alpha=0.3)
for barra in barras:
    height = barra.get_height()
    ax.text(barra.get_x() + barra.get_width()/2., height,
            f'{height:.2f}s', ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()

# Seleccionar el mejor modelo
mejor_kernel = max(resultados_kernels, key=lambda k: resultados_kernels[k]['accuracy'])
print(f"\n✓ MEJOR KERNEL: {mejor_kernel.upper()}")
print(f"  - Accuracy en Test: {resultados_kernels[mejor_kernel]['accuracy']:.4f}")
print(f"  - Mejores hiperparámetros: {mejores_params_kernels[mejor_kernel]}")

# Asignar el mejor modelo y predicciones
best_model = resultados_kernels[mejor_kernel]['modelo']
y_pred_svm = resultados_kernels[mejor_kernel]['y_pred']

# ----------------------------------------------------------------------------
# 4.3: EVALUACIÓN EXHAUSTIVA DEL MEJOR MODELO
# ----------------------------------------------------------------------------
print("\n[4.3: EVALUACIÓN EXHAUSTIVA DEL MEJOR MODELO SVM]")
print("="*80)

# Métricas del mejor modelo (ya calculadas)
acc_svm = resultados_kernels[mejor_kernel]['accuracy']
precision_svm = resultados_kernels[mejor_kernel]['precision']
recall_svm = resultados_kernels[mejor_kernel]['recall']
f1_svm = resultados_kernels[mejor_kernel]['f1_score']

print(f"\n✓ RESUMEN DEL MEJOR MODELO ({mejor_kernel.upper()}):")
print(f"  - Kernel: {mejor_kernel}")
print(f"  - Hiperparámetros: {mejores_params_kernels[mejor_kernel]}")
print(f"  - Accuracy (CV): {resultados_kernels[mejor_kernel]['cv_score']:.4f}")
print(f"  - Accuracy (Test): {acc_svm:.4f}")
print(f"  - Precision (weighted): {precision_svm:.4f}")
print(f"  - Recall (weighted): {recall_svm:.4f}")
print(f"  - F1-Score (weighted): {f1_svm:.4f}")
print(f"  - AUC Macro: {resultados_kernels[mejor_kernel]['auc_macro']:.4f}")
print(f"  - Tiempo de entrenamiento: {resultados_kernels[mejor_kernel]['tiempo_entrenamiento']:.4f}s")
print(f"  - Tiempo de predicción: {resultados_kernels[mejor_kernel]['tiempo_test']:.6f}s")

# ----------------------------------------------------------------------------
# 4.4: CURVAS ROC Y AUC (MULTICLASE)
# ----------------------------------------------------------------------------
print("\n[4.4: CURVAS ROC Y AUC]")
print("-" * 80)

# Binarizar las clases para ROC multiclase
y_test_bin = label_binarize(y_test, classes=[0, 1, 2])
n_classes = y_test_bin.shape[1]

# Obtener probabilidades
y_score = best_model.predict_proba(X_test)

# Calcular ROC y AUC para cada clase
fpr = dict()
tpr = dict()
roc_auc = dict()

for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_score[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

# Graficar ROC curves
plt.figure(figsize=(10, 8))
colors = ['blue', 'red', 'green']
especies_nombres = ['Setosa', 'Versicolor', 'Virginica']

for i, color, nombre in zip(range(n_classes), colors, especies_nombres):
    plt.plot(fpr[i], tpr[i], color=color, lw=2,
             label=f'{nombre} (AUC = {roc_auc[i]:.3f})')

plt.plot([0, 1], [0, 1], 'k--', lw=2, label='Clasificador Aleatorio')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Tasa de Falsos Positivos (FPR)', fontsize=12)
plt.ylabel('Tasa de Verdaderos Positivos (TPR)', fontsize=12)
plt.title('Curvas ROC - Clasificación Multiclase', fontsize=14, fontweight='bold')
plt.legend(loc="lower right")
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

# AUC promedio
auc_promedio = np.mean(list(roc_auc.values()))
print(f"\n✓ AUC promedio (macro): {auc_promedio:.4f}")
for i, nombre in enumerate(especies_nombres):
    print(f"  - AUC {nombre}: {roc_auc[i]:.4f}")

# ----------------------------------------------------------------------------
# 4.5: CURVA DE VALIDACIÓN (VALIDATION CURVE)
# ----------------------------------------------------------------------------
print("\n[4.5: CURVA DE VALIDACIÓN - PARÁMETRO C]")
print("-" * 80)

# Curva de validación para el parámetro C
param_range = np.logspace(-2, 3, 10)
train_scores, val_scores = validation_curve(
    best_model.named_steps['classifier'],
    best_model.named_steps['scaler'].transform(X_train),
    y_train,
    param_name='C',
    param_range=param_range,
    cv=cv_strategy,
    scoring='accuracy',
    n_jobs=-1
)

train_mean = np.mean(train_scores, axis=1)
train_std = np.std(train_scores, axis=1)
val_mean = np.mean(val_scores, axis=1)
val_std = np.std(val_scores, axis=1)

plt.figure(figsize=(10, 6))
plt.plot(param_range, train_mean, label='Score Entrenamiento', color='blue', marker='o')
plt.fill_between(param_range, train_mean - train_std, train_mean + train_std, alpha=0.2, color='blue')
plt.plot(param_range, val_mean, label='Score Validación', color='red', marker='s')
plt.fill_between(param_range, val_mean - val_std, val_mean + val_std, alpha=0.2, color='red')
plt.xscale('log')
plt.xlabel('Parámetro C', fontsize=12)
plt.ylabel('Accuracy', fontsize=12)
plt.title('Curva de Validación - Parámetro C', fontsize=14, fontweight='bold')
plt.legend(loc='best')
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

print(f"✓ Varianza en validación: {val_std.mean():.4f}")
print(f"✓ Gap train-val (sobreajuste): {(train_mean.mean() - val_mean.mean()):.4f}")

# ----------------------------------------------------------------------------
# 4.6: CURVAS DE APRENDIZAJE (LEARNING CURVES)
# ----------------------------------------------------------------------------
print("\n[4.6: CURVAS DE APRENDIZAJE]")
print("-" * 80)

train_sizes, train_scores_lc, val_scores_lc = learning_curve(
    best_model,
    X_train, y_train,
    cv=cv_strategy,
    n_jobs=-1,
    train_sizes=np.linspace(0.1, 1.0, 10),
    scoring='accuracy',
    random_state=RANDOM_STATE
)

train_mean_lc = np.mean(train_scores_lc, axis=1)
train_std_lc = np.std(train_scores_lc, axis=1)
val_mean_lc = np.mean(val_scores_lc, axis=1)
val_std_lc = np.std(val_scores_lc, axis=1)

plt.figure(figsize=(10, 6))
plt.plot(train_sizes, train_mean_lc, label='Score Entrenamiento', color='blue', marker='o')
plt.fill_between(train_sizes, train_mean_lc - train_std_lc,
                 train_mean_lc + train_std_lc, alpha=0.2, color='blue')
plt.plot(train_sizes, val_mean_lc, label='Score Validación', color='red', marker='s')
plt.fill_between(train_sizes, val_mean_lc - val_std_lc,
                 val_mean_lc + val_std_lc, alpha=0.2, color='red')
plt.xlabel('Tamaño del Conjunto de Entrenamiento', fontsize=12)
plt.ylabel('Accuracy', fontsize=12)
plt.title('Curvas de Aprendizaje', fontsize=14, fontweight='bold')
plt.legend(loc='best')
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

print(f"✓ Varianza en curvas de aprendizaje: {val_std_lc.mean():.4f}")

# ============================================================================
# FASE 5: INTERPRETACIÓN Y EVALUACIÓN FINAL
# ============================================================================
print("\n[FASE 5: INTERPRETACIÓN Y EVALUACIÓN FINAL]")
print("="*80)

# Comparación Baseline vs SVM
print("\n✓ COMPARACIÓN DE MODELOS:")
print("\n" + "-"*60)
print(f"{'Modelo':<25} {'Accuracy':<12} {'Precision':<12} {'F1-Score':<12}")
print("-"*60)
print(f"{'Baseline (Log. Reg.)':<25} {acc_baseline:<12.4f} {precision_baseline:<12.4f} {f1_baseline:<12.4f}")
print(f"{'SVM (Mejor)':<25} {acc_svm:<12.4f} {precision_svm:<12.4f} {f1_svm:<12.4f}")
print("-"*60)
print(f"{'Mejora Absoluta':<25} {(acc_svm - acc_baseline):<12.4f} {(precision_svm - precision_baseline):<12.4f} {(f1_svm - f1_baseline):<12.4f}")
print(f"{'Mejora Relativa (%)':<25} {((acc_svm/acc_baseline - 1)*100):<12.2f} {((precision_svm/precision_baseline - 1)*100):<12.2f} {((f1_svm/f1_baseline - 1)*100):<12.2f}")
print("-"*60)

# Análisis de varianza y sobreajuste
print("\n✓ ANÁLISIS DE VARIANZA Y SOBREAJUSTE:")
cv_scores = cross_val_score(best_model, X_train, y_train, cv=cv_strategy, scoring='accuracy')
print(f"  - Score CV (train): {cv_scores.mean():.4f} ± {cv_scores.std():.4f}")
print(f"  - Score Test: {acc_svm:.4f}")
print(f"  - Gap (train-test): {(cv_scores.mean() - acc_svm):.4f}")
print(f"  - Desviación estándar CV: {cv_scores.std():.4f}")

if cv_scores.std() < 0.05:
    print("  ✓ Baja varianza → Modelo estable")
else:
    print("  ⚠ Alta varianza → Revisar overfitting")

if abs(cv_scores.mean() - acc_svm) < 0.05:
    print("  ✓ Gap bajo → Sin sobreajuste significativo")
else:
    print("  ⚠ Gap alto → Posible sobreajuste")

# ============================================================================
# EJEMPLO DE PREDICCIÓN CON NUEVOS DATOS
# ============================================================================
print("\n[EJEMPLO DE PREDICCIÓN]")
print("-" * 80)

# Mapeo inverso de clases
especies_map_inv = {0: 'Iris-setosa', 1: 'Iris-versicolor', 2: 'Iris-virginica'}

# Nueva muestra
nueva_muestra = [[5.9, 3.0, 5.1, 1.8]]
prediccion = best_model.predict(nueva_muestra)[0]
probabilidades = best_model.predict_proba(nueva_muestra)[0]

print(f"\n✓ Muestra de entrada: {nueva_muestra[0]}")
print(f"✓ Especie predicha: {especies_map_inv[prediccion]}")
print(f"✓ Probabilidades:")
for i, prob in enumerate(probabilidades):
    print(f"  - {especies_map_inv[i]}: {prob:.4f}")

print("\n" + "="*80)
print("ANÁLISIS COMPLETADO")
print("="*80)
```
